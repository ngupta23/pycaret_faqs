```python
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment

data = get_data("airline")
```

```python
exp = TSForecastingExperiment()
exp.setup(data=data, fh=12, session_id=42, fig_kwargs={"renderer": "colab"})
```

```python
best3 = exp.compare_models(n_select=3)
```

```python
tuned3 = [exp.tune_model(model) for model in best3]
```

```python
exp.plot_model(tuned3)
```
