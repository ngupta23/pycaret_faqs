```python
import numpy as np
import pandas as pd
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment
```

```python
#### Intermittent Demand ----
from scipy.stats import poisson
np.random.seed(42)
data = pd.DataFrame({"demand": poisson.rvs(mu=0.25, size=144)})
```

```python
#### Create Time Series Forecasting Experiment ----
exp = TSForecastingExperiment()
global_plot_settings = {"renderer": "colab"}
exp.setup(data=data, seasonal_period=1, fh=12, fig_kwargs=global_plot_settings, session_id=42)
```

```python
#### EDA ----
exp.plot_model(plot="diagnostics", fig_kwargs={"height": 600, "width": 1200})
```

```python
#### Croston Model for intermittent demand ----
model = exp.create_model("croston")
exp.plot_model(model)
```

```python
#### Croston compared to other models ----
best = exp.compare_models(sort="RMSSE")
```
